```python
import networkx as nx
```

## Creating a graph

```python
G = nx.Graph()
```

## Nodes
Adding a single node

```python
G.add_node(1)
```

Adding multiple nodes

```python
G.add_nodes_from([2, 3, 4])
```

Looking at our added nodes

```python
G.nodes()
```

Adding a property to a node

```python
G.nodes()[1]['name'] = 'Spiderman'
G.nodes()[1]
```

## Edges
Adding a single edge between two nodes. Use the node IDs.

```python
G.add_edge(1, 2)
```

Adding multiple edges

```python
G.add_edges_from([(2, 3), (4, 1)])
```

Looking at the edges

```python
G.edges()
```

## Some Statistics

```python
print(G)
```

Get the degree of a node

```python
G.degree(2)
```

The neighbours of node 1.

```python
G[1]
```

Getting a full description of all nodes

```python
G.nodes.data()
```

## Visualise

```python
import matplotlib.pyplot as plt
```

```python
plt.clf()
nx.draw(G)
plt.show()
```

## Graph Generators

### Complete Graph

```python
complete = nx.complete_graph(10)
```

```python
plt.clf()
nx.draw(complete)
plt.show()
```

### Cycle Graph

```python
cycle = nx.cycle_graph(15)
```

```python
plt.clf()
nx.draw(cycle)
plt.show()
```

## Merging Graphs

```python
merged_graph = nx.Graph()
merged_graph.add_nodes_from(nx.cycle_graph(20))
merged_graph.add_edges_from(cycle.edges())
```

```python
plt.clf()
nx.draw(merged_graph)
plt.show()
```

## Connected Components

```python
for comp in nx.connected_components(merged_graph):
    print(comp)
```
